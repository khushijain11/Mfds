```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Function to convert 'K' and 'M' to numeric values
def convert_to_numeric(value):
    if 'K' in value:
        return float(value.replace('K', '')) * 1_000
    elif 'M' in value:
        return float(value.replace('M', '')) * 1_000_000
    else:
        return float(value)

# Load the dataset
df = pd.read_csv("C:/Users/hp/OneDrive/Documents/Mca/1 sem/mfds/social media influencers - tiktok.csv")

# Apply the conversion function
df['Likes'] = df['Likes'].apply(convert_to_numeric)
df['Shares'] = df['Shares'].apply(convert_to_numeric)

# Handling missing values
df = df.dropna(subset=['Likes', 'Shares'])

# Extract the 'Likes' and 'Shares' columns
likes = df['Likes']
shares = df['Shares']

# Calculate skewness and kurtosis for likes
likes_skewness = skew(likes)
likes_kurtosis = kurtosis(likes)

# Calculate skewness and kurtosis for shares
shares_skewness = skew(shares)
shares_kurtosis = kurtosis(shares)

# Print skewness and kurtosis values
print(f'Likes Skewness: {likes_skewness}')
print(f'Likes Kurtosis: {likes_kurtosis}')
print(f'Shares Skewness: {shares_skewness}')
print(f'Shares Kurtosis: {shares_kurtosis}')

# Plot histogram for likes
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
sns.histplot(likes, kde=True)
plt.title('Likes Distribution')

# Plot histogram for shares
plt.subplot(1, 2, 2)
sns.histplot(shares, kde=True)
plt.title('Shares Distribution')

# Show the plots
plt.tight_layout()
plt.show()
```
